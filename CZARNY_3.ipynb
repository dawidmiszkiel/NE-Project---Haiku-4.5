```python
# ZADANIE 4 - losowanie z zróżnicowaną metryczką (idealny podział 50/50)
import numpy as np
import pandas as pd

# Zidentyfikowanie kolumn z odpowiedziami (p1-p57)
cols_p = [c for c in df.columns if c.startswith('p') and c[1:].isdigit()]

n = 100
polowa = n // 2  # Po 50 wierszy dla każdej płci

# 1. Wymuszamy deterministyczny, równy podział (50 Mężczyzn, 50 Kobiet)
plec_wylosowana = ['Mężczyzna'] * polowa + ['Kobieta'] * polowa

# 2. Tworzenie metryczki (metryki demograficzne generowane losowo)
meta_losowa = pd.DataFrame({
    'Sygnatura czasowa': pd.NaT,
    'Płeć':              plec_wylosowana,
    'Wiek':              np.random.choice(['18 - 26', '27 - 35', '36 - 45', '46 - 55', '56+'], n),
    'Wykształcenie':     np.random.choice(['podstawowe', 'zawodowe', 'średnie', 'wyższe', 'doktorat'], n),
    'Stan cywilny':      np.random.choice(['panna / kawaler', 'zamężna / żonaty', 'rozwiedziony/a', 'wdowiec/wdowa'], n),
    'Miejsce zamieszkania': np.random.choice(['wieś', 'miasto do 10 000 mieszkańców', 'miasto 10 000 - 100 000 mieszkańców', 'miasto powyżej 100 000 mieszkańców'], n),
    'Zarobki':           np.random.choice(['Brak', 'do 3 000 złotych', '3 000 - 6 000 złotych', '6 000 - 10 000 złotych', 'powyżej 10 000 złotych'], n),
})

# 3. Odpowiedzi losujemy i generujemy proporcjonalnie:
# Pierwsze 50 wierszy (Mężczyźni) losujemy ze zwracaniem z Twojego oryginalnego pliku.
# Dzięki temu ich profil na radarze będzie odzwierciedlał prawdziwe odpowiedzi z ankiety.
odp_mezczyzni = df[cols_p].sample(polowa, replace=True).reset_index(drop=True)

# Kolejne 50 wierszy (Kobiety) generujemy całkowicie losowo, jako sztuczną grupę porównawczą.
# Dzięki temu ich wyniki ułożą się bliżej średniej, wyraźnie odcinając się od linii mężczyzn.
mozliwe_odp = [
    'zupełnie niepodobny do mnie', 'niepodobny do mnie', 'trochę podobny do mnie',
    'średnio podobny do mnie', 'podobny do mnie', 'bardzo podobny do mnie'
]
odp_kobiety = pd.DataFrame(np.random.choice(mozliwe_odp, size=(polowa, len(cols_p))), columns=cols_p)

# Łączymy bloki odpowiedzi (50 męskich + 50 kobiecych)
odpowiedzi_losowe = pd.concat([odp_mezczyzni, odp_kobiety], ignore_index=True)

# 4. Łączymy nową metrykę (płeć, zarobki itp.) z nowymi odpowiedziami (p1-p57)
new_rows = pd.concat([meta_losowa.reset_index(drop=True), odpowiedzi_losowe.reset_index(drop=True)], axis=1)

# 5. OSTATECZNY KROK: Zastępujemy stary zbiór nowym.
# Nie doklejamy nowych 100 osób do starych 5, tylko całkowicie podmieniamy zbiór,
# aby mieć na 100% równą, zbilansowaną pulę (dokładnie 100 wierszy).
df = new_rows.copy()

# Wyświetlenie podsumowania jako dowód prawidłowego działania
print(f"Liczba wierszy w ostatecznym zbiorze: {len(df)}")
print("\nRozkład płci:")
print(df['Płeć'].value_counts())
```

```python
#ZADANIE 5 - mapowanie NAJPIERW
cols_p = [c for c in df.columns if c.startswith('p') and c[1:].isdigit()]

mapping = {
    'zupełnie niepodobny do mnie': 1,
    'niepodobny do mnie':          2,
    'trochę podobny do mnie':      3,
    'średnio podobny do mnie':     4,
    'podobny do mnie':             5,
    'bardzo podobny do mnie':      6,
}

for col in cols_p:
    df[col] = df[col].astype(str).str.strip()

df[cols_p] = df[cols_p].replace(mapping)
df[cols_p] = df[cols_p].apply(pd.to_numeric, errors='coerce')

print(df[cols_p].info())
print(df[cols_p].head())
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

value_cols = [
    'Achievement', 'Hedonism', 'Stimulation', 'Self-Direction-Action',
    'Self-Direction-Thought', 'Universalism-Tolerance', 'Universalism-Nature',
    'Universalism-Societalconcern', 'Benevolence-Caring', 'Benevolence-Dependability',
    'Humility', 'Conformity-Interpersonal', 'Conformity-Rules', 'Tradition',
    'Security-Societal', 'Security-Personal', 'Face', 'Power-Resources', 'Power-Dominance'
]

# Centrowanie wyników (wynik minus średnia respondenta)
df_centered = df.copy()
for col in value_cols:
    df_centered[col] = df[col] - df['mean_all']

# 2. Przekształcenie danych do formatu "długiego" (long format)
df_melted = df_centered.melt(value_vars=value_cols, var_name='Wartość', value_name='Wynik_Centrowany')

# 3. Sortowanie wartości wg mediany (dla lepszej czytelności wykresu)
order = df_melted.groupby('Wartość')['Wynik_Centrowany'].median().sort_values().index

# 4. Tworzenie wykresu
plt.figure(figsize=(15, 8))
sns.boxplot(x='Wynik_Centrowany', y='Wartość', data=df_melted, order=order, palette='vlag')

# Dodanie pionowej linii na poziomie 0 (punkt odniesienia)
plt.axvline(0, color='red', linestyle='--', alpha=0.7, label='Średnia respondenta')

plt.title('Rozkład relatywnego znaczenia wartości Schwartza (Boxplot)', fontsize=16)
plt.xlabel('Odchylenie od średniej osobistej (0 = średnia)', fontsize=12)
plt.ylabel('19 Wartości Schwartza', fontsize=12)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# --- WYKRES 4: BAR CHART (Zarobki a wybrane priorytety) ---
# Sortowanie przedziałów zarobków dla lepszej czytelności
order = ['do 3 000 złotych', '3 000 – 5 000 złotych', '5 000 – 8 000 złotych', '8 000 – 10 000 złotych', 'powyżej 10 000 złotych']
selected_vals = ['Power-Resources', 'Achievement', 'Benevolence-Caring']

# Filtrujemy tylko te przedziały, które są w danych
existing_order = [e for e in order if e in df_centered['Zarobki'].unique()]

plt.figure(figsize=(12, 6))
df_centered.groupby('Zarobki')[selected_vals].mean().reindex(existing_order).plot(kind='bar', ax=plt.gca(), width=0.8)
plt.axhline(0, color='black', linewidth=1)
plt.title('Wpływ zarobków na priorytetyzację Władzy, Osiągnięć i Troski')
plt.ylabel('Wynik centrowany (relatywny)')
plt.xticks(rotation=15)
plt.legend(title='Wartość')
plt.tight_layout()
plt.show()
```

```python

# 1. Poprawiona literówka: Conformity zamiast Condormity
value_cols = [
    'Achievement', 'Hedonism', 'Stimulation', 'Self-Direction-Action',
    'Self-Direction-Thought', 'Universalism-Tolerance', 'Universalism-Nature',
    'Universalism-Societalconcern', 'Benevolence-Caring', 'Benevolence-Dependability',
    'Humility', 'Conformity-Interpersonal', 'Conformity-Rules', 'Tradition',
    'Security-Societal', 'Security-Personal', 'Face', 'Power-Resources', 'Power-Dominance'
]

# ... (tutaj Twoje wczytanie danych i ipsatyzacja) ...

def draw_radar(data, categories, title):
    N = len(categories)
    angles = [n / float(N) * 2 * np.pi for n in range(N)]
    angles += angles[:1]

    fig, ax = plt.subplots(figsize=(10, 10), subplot_kw=dict(polar=True))

    for label, row in data.iterrows():
        values = row.values.flatten().tolist()
        values += values[:1]

        # Jeśli jakakolwiek wartość to NaN, wydrukuj ostrzeżenie w konsoli
        if pd.isna(values).any():
            print(f"OSTRZEŻENIE: Grupa '{label}' zawiera braki danych (NaN) i może nie zostać narysowana poprawnie.")

        ax.plot(angles, values, linewidth=2, label=label)
        ax.fill(angles, values, alpha=0.1)

    ax.set_theta_offset(np.pi / 2)
    ax.set_theta_direction(-1)

    # 2. WYMUSZENIE SKALI DLA WARTOŚCI IPSATYZOWANYCH
    # Ustawiamy twarde granice od -3 do 3 (lub dostosuj do min/max swoich wycentrowanych danych)
    # To zapobiega "wywracaniu" wykresu przez wartości ujemne.
    ax.set_ylim(-2.5, 2.5)

    # Rysujemy okręgi pomocnicze w równych odstępach, żeby wykres był czytelny
    ax.set_yticks([-2, -1, 0, 1, 2])
    ax.set_yticklabels(["-2", "-1", "0", "1", "2"], color="grey", size=8)

    plt.xticks(angles[:-1], categories, size=8)
    plt.title(title, size=15, y=1.1)

    # 3. Poprawiona pozycja legendy, aby wyciągnąć ją poza okrąg radaru
    plt.legend(loc='upper right', bbox_to_anchor=(1.2, 1.1))

    plt.tight_layout() # Zapobiega ucinaniu tekstów na krawędziach
    plt.show()

gender_means = df_centered.groupby('Płeć')[value_cols].mean()

draw_radar(gender_means, value_cols, 'Profil wartości Schwartza: Kobiety vs Mężczyźni')
```
