```python
#ZADANIE 2
import pandas as pd
import numpy as np

oryginalny_plik = 'wyniki ankieta.csv'
df = pd.read_csv(oryginalny_plik, encoding='cp1250', sep=';')

print("Dane wczytane poprawnie!")
```

```python
#ZADANIE 2,3
nowe_kolumny = [f'p{i}' for i in range(1, len(df.columns) + 1)]
df.columns = nowe_kolumny

print("Nazwy kolumn po zmianie:")
print(df.columns.tolist())
```

```python
#ZADANIE 4
liczba_nowych_wierszy = 100

losowe_wiersze = df.sample(n=liczba_nowych_wierszy, replace=True, random_state=42)
df = pd.concat([df, losowe_wiersze], ignore_index=True)

print(f"Dodano {liczba_nowych_wierszy} wierszy. Nowy rozmiar tabeli to: {df.shape}")
```

```python
#ZADANIE 5
nazwa_koloru_grupy = 'czarny'
nazwa_nowego_pliku = f'{nazwa_koloru_grupy}.csv'

df.to_csv(nazwa_nowego_pliku, index=False)
print(f"Zapisano gotowy plik jako: {nazwa_nowego_pliku}")
```

```python
#ZADANIE 6
print("--- 1. df.head() - Podgląd pierwszych 5 wierszy ---")
display(df.head())

print("\n--- 2. df.tail() - Podgląd ostatnich 5 wierszy ---")
display(df.tail())

print("\n--- 3. df.sample(5) - Losowanie 5 wierszy z tabeli ---")
display(df.sample(5))

print("\n--- 4. df.info() - Typy danych, brakujące wartości, zużycie pamięci ---")
df.info()

print("\n--- 5. df.describe() - Statystyki opisowe dla kolumn numerycznych ---")
display(df.describe())

print("\n--- 6. df.nunique() - Liczba unikatowych wartości w kolumnach ---")
display(df.nunique())

print("\n--- 7. df['p1'].value_counts() - Częstotliwość występowania wartości ---")
display(df['p1'].value_counts())
```
